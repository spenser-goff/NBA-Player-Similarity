```python
import requests
import unidecode
import csv
import pandas as pd
from bs4 import BeautifulSoup
from bs4 import Comment
```

```python
alpha_lst = ['a', 'b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','y','z']
url = 'https://www.basketball-reference.com/players/{}/'
playerpage_lst = []
playername_lst = []
playeryear_lst = []
for letter in alpha_lst:
    page_link = 'https://www.basketball-reference.com'
    page_url = url.format(letter)
    r = requests.get(page_url)
    soup = BeautifulSoup(r.content)
    playerpage = ''
    for a in soup.find('div', {'id':'content'}).find_all('td', {'data-stat':'year_max'}):
        playeryear_lst.append(a.get_text())
    for a in soup.find('div', {'id':'content'}).find_all('a', href=True):
        if "/players/" in a['href']:
            name = str(a.get_text())
            name = unidecode.unidecode(name)
            name = name.lower()
            playername_lst.append(name)
            playerpage_lst.append(page_link + str(a['href']))
```

```python
df = pd.DataFrame(columns=['Name', 'Page Link', 'Last Year'])
df['Name'] = playername_lst
df['Page Link'] = playerpage_lst
df['Last Year'] = playeryear_lst
df
```

```python
df.to_csv('player_info.csv', index=False)
```
