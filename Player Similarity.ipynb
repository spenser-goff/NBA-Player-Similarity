```python
player_info = player_info.iloc[::-1]
player_info = player_info.reset_index(drop=True)
player_info['AST%'] = player_info['AST%'].apply(lambda x: x/100)
player_info['STL%'] = player_info['STL%'].apply(lambda x: x/100)
player_info['BLK%'] = player_info['BLK%'].apply(lambda x: x/100)
player_info['TRB%'] = player_info['TRB%'].apply(lambda x: x/100)
player_info['TOV%'] = player_info['TOV%'].apply(lambda x: x/100)
player_info['USG%'] = player_info['USG%'].apply(lambda x: x/100)
player_info
```

```python
df = player_info.sample(n=2000)

names = df['Player']

df = df.drop(columns=['Player', 'PG %', 'SG %', 'SF %', 'PF %', 'C %'])
df = df.reset_index(drop=True)
length = len(names)
zeros = np.zeros((length, length))
sim_matrix = pd.DataFrame(zeros, columns=names)
for i, r1 in df.iterrows():
    for j, r2 in df.iterrows():
        if i != j:
            sim_matrix.iat[i, j] = np.dot(r1, r2)/(norm(r1)*norm(r2))
        else:
            sim_matrix.iat[i, j] = 1
sim_matrix.index = names
sim_matrix
```
